# 1. Introduction

**Background**

In 2020, as we may see that Net Promotor Score (NPS) is kind of new to be implemented as one of the aspect of measuring training delivery. Basically learners has been asked about on the scale of 1-10 how they recommend this training to others. 

NPS itself would be very useful for Academy as their references to do such things as :
1. Predicting whether the training should be re-deliver again for the next batch
2. Evaluating about the training delivery based on learners recommendation

The problem is the incompleteness in the NPS data due to many causes such as learners forgot to fill the feedback survey that contains NPS question. Due to that condition, Academy finds it hard to decide whether they should re-deliver the training for another batch or not. It would be very helpful if we can help Academy to predict the NPS of the training that has been delivered and has been not delivered before.

**Data Sources**

The dataset that will be used is Training Feedback Data 2020 from Telkom Data Bay. The total of the data is 1581 rows that will be divided into 70% train data and 30% test data. The result of this project is for training purpose only.

**Objectives**

The aim of this project is to create a machine learning model using *classification* to determine whether NPS of the training will be above or under +40 based on features such as training category, city, academy, training type, and provider_category.
The purpose of set up NPS +40 is because the current standard of NPS Training is +40.

# 4. Data Preprocessing

```python
#Copy data
training_data = training.copy()
```

```python
#Drop column that will not be used in this project
training_data.drop(['Tipe'], inplace=True, axis=1)
```

```python
#Manipulating columns by renaming column name
training_data = training_data.rename(
    columns={"Nama Pelatihan" : "training_name", 
             "Objid Pelatihan" : "objid",
             "Category" : "category",
             "Lokasi Pelatihan" : "city",
             "Provider" : "provider",
             "Provider Category" : "provider_category",
             "Academy Event" : "academy",
             "Location" : "location",
             "Event Type" : "training_type",
             "Status" : "status",
             "Tgl Mulai" : "start_date",
             "Tgl Selesai" : "end_date",
             "JML Peserta" : "total_participant",
             "JML Confirmed" : "total_confirmed",
             "JML Peserta Hadir" : "total_attendance",
             "JML UBPP Inst" : "total_feed_instructor",
             "JML UBPP Delivery" : "total_feed_delivery",
             "UBPP Inst" : "feed_instructor",
             "UBPP Akom" : "feed_acomodation",
             "UBPP Sarana" : "feed_facility",
             "UBPP Laborat" : "feed_lab",
             "UBPP Penyelenggaraan" : "feed_delivery",
             "UBPP Kafetaria" : "feed_cafe",
             "UBPP Materi" : "feed_material",
             "Net Promotor Score" : "nps",
             "Customer Satisfaction" : "cs",
             "First Response & Average Handling" : "frah",
             "Customer Effort Score" : "ces"           
             }
)
```

```python
#View the top 5 rows of the dataset using the head() function
training_data.head(5)
```

```python
#Fill missing value in column feed_instructor with data mean
training_data['feed_instructor'].fillna((training_data['feed_instructor'].mean()), inplace=True)
training_data.info()
```

```python
#Remove missing values
training_data = training_data.dropna()
training_data.info()
```

```python
#Specifying data types
training_data['start_date'] = pd.to_datetime(training_data['start_date'])
training_data['end_date'] = pd.to_datetime(training_data['end_date'])
training_data.info()
```

```python
training_data.isnull().sum()
```


Next, I want to add new column along with labeling NPS with current standard used in this company. For your information, the NPS standard for training is **+40** so that I will divide NPS Category into two :
1. Under 40, means the NPS result is considered below the current standard
2. Above 40, means the NPS result is considered above the current standard

```python
#create a new column named nps_category and fill it with blank first
training_data['nps_category'] = ''
```

```python
#create a list of our condition of NPS category as stated before with +40 as the standard
conditions = [
    (training_data['nps'] < 40),
    (training_data['nps'] >= 40)
]

#create a list of values we want to assign for each condition, this list is flexible to extend
values = ['Under 40', 'Above 40']

#use np.select to assign values to it using our list as arguments
training_data['nps_category'] = np.select(conditions, values)

#display updated DataFrame
training_data.head()
```

```python
#initialize label encode
object_cols = ['category', 'city', 'academy','training_type','provider_category','nps_category']

label_encoder = LabelEncoder()
for col in object_cols:
    training_data[col] = label_encoder.fit_transform(training_data[col])
training_data.head()
```

```python
#Defining features that will be used in model definition
features = ['category', 'city', 'academy','training_type','provider_category']

x = training_data[features]
y = training_data['nps_category']
```

Next step is spliting the data into two cluster : training set and test set randomly. I will use training set to train the model follows by evaluating its performance with test set. Thus, we will use train_test_split() with composition is 70% training set and 30% test set.

```python
#split data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

# 5. Model Definition

The model that will be used in this project are model related with Classification, which are :
1. Decision Tree
2. Random Forest
3. Logistic Regression

# 6. Model Training

## 6.1 Decision Tree

```python
dtree = DecisionTreeClassifier(criterion='entropy')
dtree.fit(x_train, y_train)

y_pred = dtree.predict(x_test)

print("Accuracy Score:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## 6.2 Random Forest

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(x_train, y_train)

y_pred = random_forest.predict(x_test)

print("Accuracy Score:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## 6.3 Logistic Regression

```python
LogR = LogisticRegression(solver='lbfgs')
LogR.fit(x_train, y_train)

y_pred = LogR.predict(x_test)

print("Accuracy Score:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
